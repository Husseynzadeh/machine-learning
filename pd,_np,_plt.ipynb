# Mini Batch Gradient Descent

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import add_dummy_feature
np.random.seed(42)
m = 100
X = 2*np.random.rand(m,1)
y = 4 + 3*X + np.random.randn(m,1)
X_b = add_dummy_feature(X)
```

```python
n_epochs = 100
m = len(X_b)
theta = np.random.randn(2,1)
batch_size = 10
eta = 0.1
n_batches = int(m/batch_size)
```

```python
for epoch in range(n_epochs):
  for batch in range(n_batches):
      random_index = np.random.randint(m)
      xi = X_b[random_index:random_index+1]
      yi = y[random_index:random_index+1]
      gradients = xi.T @(xi @ theta - yi)
      theta = theta - eta* gradients
print(theta)
```

```python
n_epochs = 470
m = len(X_b)
theta = np.random.randn(2,1)
batch_size = 10
eta = 0.1
n_batches = m//batch_size
for epoch in range(n_epochs):
  for batch in range(n_batches):
      random_indices = np.random.randint(0, m, batch_size)
      X_batch = X_b[random_indices]
      y_batch = y[random_indices]
      gradients = X_batch.T @(X_batch @ theta - y_batch)
      theta = theta - eta* gradients
print(theta)
```

```python
# Initialize parameters
theta = np.random.randn(2, 1)  # 2 features: 1 for bias, 1 for X
batch_size = 10  # Mini-batch size
eta = 0.1  # Learning rate
n_epochs = 100  # Number of epochs

# Number of mini-batches per epoch
n_batches = m // batch_size

# Mini-batch gradient descent loop
for epoch in range(n_epochs):
    for batch in range(n_batches):
        # Randomly select indices for the mini-batch
        random_indices = np.random.randint(0, m, batch_size)

        # Select the mini-batch of features and labels
        X_batch = X_b[random_indices]
        y_batch = y[random_indices]

        # Compute predictions
        predictions = X_batch @ theta

        # Compute the gradient for the mini-batch
        gradients = X_batch.T @ (predictions - y_batch) / batch_size

        # Update parameters (theta)
        theta -= eta * gradients


# Final parameters after training
print(theta)
```
