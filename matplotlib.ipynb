```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("hmavrodiev/london-bike-sharing-dataset")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv(path + "/london_merged.csv")
df
```

```python
df['timestamp']= pd.to_datetime(df['timestamp'])
df.head()
```

```python
w = df.groupby(['is_weekend'])['cnt'].mean()
plt.bar(['weekday', 'weekend'], w.values)
```

```python
h = df.groupby(['is_holiday'])['cnt'].mean()
plt.bar(['holiday', 'not holiday'], h.values)
```

```python
plt.figure(figsize = (14,7))
df.groupby(df['timestamp'].dt.hour)['cnt'].mean().plot(kind = 'bar')
```

```python
plt.figure(figsize = (10,5))
df.groupby(df['timestamp'].dt.month)['cnt'].mean().plot(kind = 'bar')
```

```python
plt.figure(figsize = (10,5))
df.groupby(df['timestamp'].dt.dayofweek)['cnt'].mean().plot(kind = 'bar')
```

```python
s = df.groupby(df['season'])['cnt'].mean()
plt.bar(['spring', 'summer','autumn','winter'],s.values);
```

```python
plt.figure(figsize = (10,5))
df.groupby(df['season'])['cnt'].mean().plot(kind = 'pie')
```

```python
plt.pie(s.values, labels = ['spring', 'summer', 'autumn', 'winter'], autopct = '%.1f%%');
```

```python
import seaborn as sns
c = df[['t1','hum', 'wind_speed', 'cnt']].corr()
sns.heatmap(c, annot = True, cmap = 'coolwarm');
```
